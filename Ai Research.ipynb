##### **Overview of Dataset:**

The dataset is structured with several key columns that capture diverse aspects of participants' perspectives. These include:

1. **Timestamp:** The timestamp of each survey response, providing temporal context to the data.

2. **Knowledge Ratings (column1):** Participants' self-assessed levels of knowledge and understanding of Artificial Intelligence, rated on a scale from 1 to 5.

3. **Personal Use Ratings (column2):** Frequency ratings, ranging from 1 to 5, indicating how often participants employ AI technologies for personal purposes.

4. **School-Related Use Ratings (column3):** Similar to personal use ratings, this column reflects participants' frequency of utilizing AI for school-related tasks.

5. **Interest in AI Career (column4):** Participants' level of interest in pursuing a career in Artificial Intelligence, rated on a scale of 1 to 5.

6. **Awareness of Chat-GPT (column5):** A binary indicator denoting whether participants are aware of Chat-GPT (Yes/No).

7. **College (column6):** The academic discipline to which participants belong, providing insights into potential variations in responses across different colleges.

##### **Objective of Analysis:**

The primary objective of this analysis on the "AI on Campus Research Survey" is to gain actionable insights into the perceptions and behaviors of participants within an academic setting regarding Artificial Intelligence (AI). Through thorough exploratory data analysis (EDA), the specific goals include:

1. **Understanding Knowledge and Usage Patterns:** Evaluate participants' self-assessed knowledge levels of AI and explore the frequency of AI usage for personal and school-related tasks.

2. **Examining Career Interest in AI:** Investigate participants' interest levels in pursuing a career in Artificial Intelligence, identifying potential trends and patterns.

3. **Assessing Awareness of Chat-GPT:** Analyze participants' awareness of Chat-GPT, differentiating responses between those who are aware and those who are not.

4. **Comparative Analysis Across Colleges:** Explore variations in responses across different academic disciplines (colleges) to identify trends or unique patterns.

5. **Temporal Analysis:** Investigate temporal trends to discern any variations in responses over different months or years.

By accomplishing these objectives, this analysis aims to provide valuable insights for decision-making and further exploration of AI-related perceptions and behaviors within the academic community.

#### *Analysing overall distribution*

```python
# Plotting the distribution
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Overall Distribution of Responses',  fontweight='semibold', fontsize=25)

# Plot 1
ax1 = sns.countplot(ax=axes[0, 0], data=df, x='On a scale from 1 to 5, how would you rate your knowledge and understanding of Artificial Intelligence (AI)?',edgecolor='black')
ax1.set_xlabel('Ratings')
for p in ax1.patches:
    ax1.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Plot 2
ax2 = sns.countplot(ax=axes[0, 1], data=df, x='On a scale from 1 to 5, how often do you use Artificial Intelligence (AI) for personal use?',edgecolor='black')
ax2.set_xlabel('Ratings')
for p in ax2.patches:
    ax2.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Plot 3
ax3 = sns.countplot(ax=axes[0, 2], data=df, x='On a scale from 1 to 5, how often do you use Artificial Intelligence (AI) for school-related tasks?',edgecolor='black')
ax3.set_xlabel('Ratings')
for p in ax3.patches:
    ax3.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Plot 4
ax4 = sns.countplot(ax=axes[1, 0], data=df, x='On a scale from 1 to 5, how interested are you in pursuing a career in Artificial Intelligence?',edgecolor='black')
ax4.set_xlabel('Ratings')
for p in ax4.patches:
    ax4.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Plot 5
ax5 = sns.countplot(ax=axes[1, 1], data=df, x='Do you know what Chat-GPT is?',edgecolor='black')
ax5.set_xlabel('Ratings')
for p in ax5.patches:
    ax5.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Plot 6
ax6 = sns.countplot(ax=axes[1, 2], data=df, x='What college are you in?', order=df['What college are you in?'].value_counts().index,edgecolor='black')
ax6.set_xlabel('Ratings');ax6.set_xticklabels(ax6.get_xticklabels(), rotation=90) 
for p in ax6.patches:
    ax6.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show plots
plt.show()
```

####  *Correlation Analysis*

```python
#creating a copy of the DataFrame so we can rename the columns and make the correlation matrix look more clear and engaging
df1 = df.copy()
df1.columns = ['Timestamp', 'column1', 'column2', 'column3', 'column4', 'column5', 'column6']
df1.head()
```

```python
correlation_cols=['column1','column2','column3','column4']
# Creating a correlation matrix
correlation_matrix = df1[correlation_cols].corr()

# Ploting the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix', fontweight='semibold', fontsize=25)
plt.show()
```

```python
# Scatter plot for personal use vs. school-related use
plt.figure(figsize=(8, 6))
sns.scatterplot(x='column2', y='column3', data=df1, hue='column6', palette='Set1', s=70)
plt.title('Personal Use vs. School-Related Use of AI', fontweight='semibold',fontsize=20)
plt.xlabel('Personal Use Ratings')
plt.ylabel('School-Related Use Ratings')
plt.legend(title='College', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.show()
```

```python
# Grouped bar plot for awareness vs. interest
plt.figure(figsize=(10, 6))
ax=sns.countplot(x='column5', hue='column4', data=df1, palette='Set3',edgecolor='black')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title('Awareness vs. Interest in AI Career', fontweight='semibold',fontsize=20)
plt.xlabel('Awareness of Chat-GPT')
plt.ylabel('Count')
plt.legend(title='Interest in AI Career', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.show()
```

```python
Columns = ['On a scale from 1 to 5, how would you rate your knowledge and understanding of Artificial Intelligence (AI)?',
            'On a scale from 1 to 5, how often do you use Artificial Intelligence (AI) for personal use?', 
            'On a scale from 1 to 5, how often do you use Artificial Intelligence (AI) for school-related tasks?',
            'On a scale from 1 to 5, how interested are you in pursuing a career in Artificial Intelligence?', 
            'Do you know what Chat-GPT is?']

for column in Columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, hue='What college are you in?', data=df, palette='muted')
    plt.title(f'Comparative Analysis: {column}',fontweight='semibold',fontsize=15)
    plt.xlabel('Ratings')
    plt.ylabel('Count')
    plt.legend(title='College', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df1.drop(['column6'], axis=1),palette='RdBu')  
plt.title('Box Plots for Each Column',fontweight='semibold',fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Extracting month and year from the timestamp
df['Month'] = df['Timestamp'].dt.month
df['Year'] = df['Timestamp'].dt.year

# Count of responses per month and year
temporal_data = df.groupby(['Year', 'Month']).size().reset_index(name='Response Count')

# Line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Response Count', hue='Year', data=temporal_data, marker='o',palette='Set1')
plt.title('Temporal Analysis: Responses Over Time', fontweight='semibold',fontsize=20)
plt.xlabel('Month')
plt.ylabel('Response Count')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
